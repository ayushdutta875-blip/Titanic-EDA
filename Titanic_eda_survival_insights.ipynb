```python
print("========== TITANIC REPORT ==========") # Final Summarize Report

print("\nTotal Passengers:", len(df))

print("\nSurvival Rate:")
print(round(df['Survived'].mean()*100,2), "%")

print("\nGender Distribution:")
print(df['Sex'].value_counts())

print("\nPassenger Class Distribution:")
print(df['Pclass'].value_counts())

print("\nAverage Age:")
print(round(df['Age'].mean(),2))

print("\nAverage Fare:")
print(round(df['Fare'].mean(),2))

print("\nSurvival Rate By Gender:")
print(df.groupby('Sex')['Survived'].mean()*100)

print("\nSurvival Rate By Passenger Class:")
print(df.groupby('Pclass')['Survived'].mean()*100)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nMost Common Title:")
print(df['Name'].mode()[0])

print("\n========== END ==========")
```

```python
# Now Time To Visualize Some Data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating a bar chart showing the number of passengers in each passenger class.
plt.figure(figsize=(8,6))
a=df['Pclass'].value_counts()
plt.bar(a.index,a.values)
plt.xlabel("Pclasses")
plt.ylabel("No of passengers")
plt.title("No of passenger in each class")
plt.show()
```

```python
# Creating a countplot showing survival distribution.
plt.figure(figsize=(8,5))
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
#Creating a countplot comparing survival by gender.
plt.figure(figsize=(8,6))
a=df.groupby('Sex')['Survived'].value_counts()
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title("Survival by Gender")

plt.show()
```

```python
# Boxplot of Age by Passenger Class
sns.boxplot(x='Pclass', y='Age', data=df)

plt.title("Age Distribution by Passenger Class")

plt.show()
```

```python
# Scatter Plot: Age vs Fare
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',
    data=df
)

plt.title("Age vs Fare")

plt.show()
```

```python
#Correlation Heatmap
corr = df[['Survived' , 'Pclass' , 'Age' , 'Fare' , 'SibSp' , 'Parch']].corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr , annot = True , cmap = 'Greens' , linewidth = 0.5)
plt.title('Correlation Heatmap Of All Features Of Titanic')
plt.show()
```
